```python
from facebook_scraper import get_profile
import pandas as pd
```

```python
def readData(fileName):
    ''' 
    Read data from the given file
    Mainly used to read post IDs list
    '''
    f = open(fileName, 'r', encoding='utf-8')
    data = []
    for i, line in enumerate(f):
        try:
            line = repr(line)
            line = line[1:len(line) - 3]
            data.append(line)
        except:
            print("Error writing line")
    return data
```

```python
post_IDs = readData('Data/user_id.csv') # Chuy·ªÉn data user_id.csv th√†nh danh s√°ch post_IDs
print(len(post_IDs))
```

*L·∫•y d·ªØ li·ªáu*

```python
list =[]
for account in post_IDs:
    try:
        profile = get_profile(account)
        list.append(profile)
        print(profile)
    except Exception as e:
        print(f"An error occurred with account {account}: {e}")
        continue
df = pd.DataFrame(list)
df.to_csv('Data/user.csv', header=False, index=False)
```
