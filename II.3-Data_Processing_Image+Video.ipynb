```python
import pandas as pd
import matplotlib.pyplot as plt

# Tính toán số lượng cho từng loại
video_and_image = df[(df['has_video'] == 1) & (df['has_image'] == 1)].shape[0]
video_only = df[(df['has_video'] == 1) & (df['has_image'] == 0)].shape[0]
image_only = df[(df['has_video'] == 0) & (df['has_image'] == 1)].shape[0]
none = df[(df['has_video'] == 0) & (df['has_image'] == 0)].shape[0]

# Vẽ biểu đồ
labels = ['Video và Hình ảnh', 'Chỉ Video', 'Chỉ Hình ảnh', 'Không có gì']
sizes = [video_and_image, video_only, image_only, none]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  # Đảm bảo rằng biểu đồ được vẽ là một hình tròn.
plt.show()
```
