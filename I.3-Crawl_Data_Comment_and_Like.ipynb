```python
def readData(fileName):
    ''' 
    Read data from the given file
    Mainly used to read post IDs list
    '''
    f = open(fileName, 'r', encoding='utf-8')
    data = []
    for i, line in enumerate(f):
        try:
            line = repr(line)
            line = line[1:len(line) - 3]
            data.append(line)
        except:
            print("Error writing line")
    return data
```
