```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/data_user.csv')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df
```

```python
def convert_name_to_live_in(df):
    for phrase in ["H·ªçc vi·ªán","ƒê·∫°i H·ªçc","Dai hoc","ƒê·∫°i hoc","ƒë·∫°i h·ªçc","trung c·∫•p","Trung c·∫•p","Trung C·∫•p", "trung h·ªçc","Cao ƒë·∫≥ng","ƒê·∫°i h·ªçc","ti·∫øn s·ªπ","th·∫°c s·ªπ","Ti·∫øn s·ªπ","THPT","THCS","Trung h·ªçc","Trung hoc","Trung Hoc"]:
        df.loc[df['Name'].str.contains(phrase, na=False), 'H·ªçc v·∫•n'] = df['N∆°i t·ª´ng s·ªëng']
    for phrase in ["Qu√™ qu√°n", "Th√†nh ph·ªë hi·ªán t·∫°i"]:
        df.loc[df['Name'].str.contains(phrase, na=False), 'N∆°i t·ª´ng s·ªëng'] = df['Name']
    return df
# S·ª≠ d·ª•ng h√†m
df = convert_name_to_live_in(df)
df['N∆°i t·ª´ng s·ªëng'] = df['N∆°i t·ª´ng s·ªëng'].apply(lambda x: np.nan if ("Qu√™ qu√°n" not in str(x)) and ("Th√†nh ph·ªë hi·ªán t·∫°i" not in str(x)) else x)

df
```

```python
# Ch·ªâ gi·ªØ l·∫°i c√°c c·ªôt c·∫ßn thi·∫øt,ta s·∫Ω ph√¢n t√≠ch name ·ªü ph·∫ßn d·ªØ li·ªáu kh√°c(d·ªØ li·ªáu g·ªìm comments_pro v√† likes_pro) ƒë·∫ßy ƒë·ªß h∆°n
filtered_df = df[['id', 'N∆°i t·ª´ng s·ªëng']] 

filtered_df
```

```python
# X√≥a c√°c d√≤ng tr√πng l·∫∑p d·ª±a tr√™n tr∆∞·ªùng 'id'
filtered_df = filtered_df.drop_duplicates(subset='id')

# X√≥a c√°c d√≤ng ch·ª©a gi√° tr·ªã NaN trong tr∆∞·ªùng 'id'
filtered_df = filtered_df[filtered_df['id'].notna()]
filtered_df = filtered_df.reset_index(drop=True)
# L∆∞u DataFrame v√†o file CSV
filtered_df.to_csv('Data/filtered_data.csv', index=False)
filtered_df
```
