*Import 2 th∆∞ vi·ªán pandas v√† numpy ƒë·ªÉ x·ª≠ l√Ω*

```python
import pandas as pd
import numpy as np
import ast
```

*ƒê·ªçc d·ªØ li·ªáu t·ª´ file chuaxuly.csv chuy·ªÉn th√†nh ƒë·ªãnh d·∫°ng dataframe*

```python
df = pd.read_csv('Data/chuaxuly.csv')
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 100)
df
```

```python
df.info() #t√≥m t·∫Øt ng·∫Øn g·ªçn v·ªÅ d·ªØ li·ªáu
```

*X√≥a c√°c c·ªôt missing data ho·∫∑c c√°c c·ªôt ta kh√¥ng ph√¢n t√≠ch ƒë·∫øn*

```python
# Danh s√°ch c√°c c·ªôt ta x√≥a
columns_to_drop = ['comments_full','sharers','video_height','header','was_live','shared_text',
                   'video_duration_seconds','video_quality','video_size_MB','video_watches',
                   'video_width','is_live','factcheck','shared_post_id','shared_time','shared_user_id',
                   'shared_username','shared_post_url','available','original_text','shared_text'
                   ,'timestamp','with','video_ids','videos','w3_fb_url','reactors','fetched_time','image_id','link']

# S·ª≠ d·ª•ng ph∆∞∆°ng th·ª©c drop
df = df.drop(columns=columns_to_drop)
df.info()
```

*L·ªçc c√°c b√†i vi·∫øt ch·ªâ xu·∫•t hi·ªán sau th√°ng 10*

```python
# Chuy·ªÉn ƒë·ªïi c·ªôt 'time' th√†nh ƒë·ªãnh d·∫°ng datetime
df['time'] = pd.to_datetime(df['time'])

# L·ªçc DataFrame ƒë·ªÉ ch·ªâ gi·ªØ l·∫°i c√°c d√≤ng trong th√°ng 10
df = df[df['time'].dt.month == 10]
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.sort_values('time')
df = df.drop_duplicates(subset='post_id')
df = df.reset_index(drop=True)
df
```

```python
df.info()
```

*T√°ch c√°c d·ªØ li·ªáu t·ª´ d·ªØ li·ªáu g·ªëc ƒë·ªÉ khi ta ph√¢n t√≠ch d·ªÖ d√†ng h∆°n*

**Th√™m 2 c·ªôt th·ªÉ hi·ªán b√†i vi·∫øt c√≥ ƒëƒÉng h√¨nh ·∫£nh hay video kh√¥ng**

```python
df['image_ids'] = df['image_ids'].apply(ast.literal_eval)
df['image_ids'].apply(pd.Series)
# Th√™m c·ªôt 'has_video'
df['has_video'] = df['video_id'].apply(lambda x: 0 if np.isnan(x) else 1)
# Th√™m c·ªôt 'has_image'
df['has_image'] = df['image_ids'].apply(lambda x: 1 if (x and len(x) > 0) else 0)

```

**Th√™m c√°c c·ªôt th·ªÉ hi·ªán l∆∞·ª£t reaction b√†i vi·∫øt**

```python
df['reactions'] = df['reactions'].apply(ast.literal_eval)
df_reactions = df['reactions'].apply(pd.Series)

df_reactions = df_reactions.drop(['th√≠ch', 'y√™u th√≠ch', 'th∆∞∆°ng th∆∞∆°ng', 'bu·ªìn', 'ph·∫´n n·ªô'], axis=1)
df_reactions=df_reactions.fillna(value=0)
df = pd.concat([df, df_reactions], axis=1)
df.drop(['reactions'], axis=1)

```

```python
# T·∫°o c·ªôt 'other' b·∫±ng c√°ch c·ªông c√°c c·ªôt 'wow', 'care', 'sad', v√† 'angry'
df['other'] = df[['wow', 'care', 'sad', 'angry']].sum(axis=1)
```

```python
df.to_csv('Data/daxuly.csv', index=False)
```
