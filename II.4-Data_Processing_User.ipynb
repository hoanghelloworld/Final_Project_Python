```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer
```

## X√¢y d·ª±ng m√¥ h√¨nh d·ª± ƒëo√°n t√™n ##

```python
# ƒê·ªçc d·ªØ li·ªáu t·ª´ file CSV
data = pd.read_csv('Data/Dataset-Gender.csv')

# Chuy·ªÉn ƒë·ªïi t√™n th√†nh c√°c ƒë·∫∑c tr∆∞ng b·∫±ng c√°ch s·ª≠ d·ª•ng CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['Full_Name'])

# Chia d·ªØ li·ªáu th√†nh t·∫≠p hu·∫•n luy·ªán v√† t·∫≠p ki·ªÉm tra
X_train, X_test, y_train, y_test = train_test_split(X, data['Gender'], test_size=0.2, random_state=42)

# X√¢y d·ª±ng m√¥ h√¨nh Random Forest
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)

# ƒê√°nh gi√° ƒë·ªô ch√≠nh x√°c c·ªßa m√¥ h√¨nh tr√™n t·∫≠p ki·ªÉm tra
accuracy = model.score(X_test, y_test)

print("ƒê·ªô ch√≠nh x√°c c·ªßa m√¥ h√¨nh l√†: ", accuracy)

```

*Th·ª≠ nghi·ªám m·ªôt ch√∫t*

```python
# T√™n c·∫ßn d·ª± ƒëo√°n
name = 'Xu√¢n Th·∫Øm'

# Chuy·ªÉn ƒë·ªïi t√™n th√†nh ƒë·∫∑c tr∆∞ng
features = vectorizer.transform([name])

# D·ª± ƒëo√°n gi·ªõi t√≠nh
prediction = model.predict(features)

# In ra k·∫øt qu·∫£
if prediction[0] == 1:
    print('Gi·ªõi t√≠nh d·ª± ƒëo√°n: Nam')
else:
    print('Gi·ªõi t√≠nh d·ª± ƒëo√°n: N·ªØ')
```

*Ta th·∫•y c·ªôt profileName ·ªü df1 t∆∞∆°ng ·ª©ng v·ªõi c·ªôt name ·ªü df2 n√™n ta c·∫ßn ƒë·ªìng nh·∫•t 2 c·ªôt n√†y v√† l∆∞u v√†o df m·ªõi ch·ªâ ch·ª©a 2 c·ªôt profileUrl v√† name ƒë·ªÉ d·ªÖ d√†ng ph√¢n t√≠c*

```python
# ƒê·ªïi t√™n c·ªôt 'profileName' trong df1 th√†nh 'name'
df1 = df1.rename(columns={'profileName': 'name'})

# T·∫°o dataframe m·ªõi t·ª´ df1 v√† df2 ch·ªâ bao g·ªìm profileUrl v√† name
df = pd.concat([df1, df2], ignore_index=True)
df = df[['profileUrl', 'name']]
df = df.drop_duplicates(subset='profileUrl', keep='first')
df = df.reset_index(drop=True)
df

```

*X·ª≠ d·ª•ng m√¥ h√¨nh d·ª± ƒëo√°n gi·ªõi t√≠nh ƒë·ªÉ d·ª± ƒëo√°n t√™n v√† th√™m c·ªôt gender*

```python
# T·∫°o m·ªôt c·ªôt m·ªõi 'Gender' b·∫±ng c√°ch d·ª± ƒëo√°n gi·ªõi t√≠nh d·ª±a tr√™n t√™n
df['Gender'] = df['name'].apply(lambda name: 'Nam' if model.predict(vectorizer.transform([name]))[0] == 1 else 'N·ªØ')

# In ra DataFrame sau khi ƒë√£ th√™m c·ªôt 'Gender'
df
```

*Ph√¢n t√≠ch s·ª± t∆∞∆°ng quan gi·ªØa hai gi·ªõi t√≠nh nam v√† n·ªØ khi t∆∞∆°ng t√°c*

```python
import matplotlib.pyplot as plt

# ƒê·∫øm s·ªë l∆∞·ª£ng nam v√† n·ªØ
counts = df['Gender'].value_counts()

# V·∫Ω bi·ªÉu ƒë·ªì tr√≤n
plt.pie(counts, labels = counts.index, autopct='%1.1f%%')
plt.title('Ph√¢n b·ªë gi·ªõi t√≠nh')
plt.show()
```
